```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
from scipy.stats import sem

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_df = pd.merge(mouse_metadata, study_results, how="outer", on=["Mouse ID"])
# Display the data table for preview
combined_df
```

```python
# Checking the number of mice.
mice_total = len(combined_df["Mouse ID"].value_counts())
mice_total
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = combined_df.loc[combined_df.duplicated(subset=["Mouse ID", "Timepoint"]),"Mouse ID"].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate = combined_df.set_index('Mouse ID')
duplicate.loc['g989']
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = combined_df.loc[combined_df['Mouse ID'] != "g989"]
clean_df.head(-1)
```

```python
# Checking the number of mice in the clean DataFrame.
new_mouse_count = len(clean_df["Mouse ID"].value_counts())
new_mouse_count_df = new_mouse_count
new_mouse_count_df
```

```python
new_mouse_count_m = len(clean_df["Sex"].value_counts())
new_mouse_count_m_df = new_mouse_count_m
new_mouse_count_m_df
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
regimen_df = clean_df.groupby("Drug Regimen")

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
mean_tum = (regimen_df["Tumor Volume (mm3)"].mean())
median_tum = (regimen_df["Tumor Volume (mm3)"].median())
variance_tum = (regimen_df["Tumor Volume (mm3)"].var())
std_dev_tum = (regimen_df["Tumor Volume (mm3)"].std())
sem_tum = (regimen_df["Tumor Volume (mm3)"].sem())

# Assemble the resulting series into a single summary dataframe.
summary_df = pd.DataFrame({"Volume Mean": mean_tum,
                          "Volume Median": median_tum,
                          "Volume Variance": variance_tum,
                          "Volume Std Deviation": std_dev_tum,
                          "Volume SEM": sem_tum})
summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
regimen_df = clean_df.groupby("Drug Regimen")

# Using the aggregation method, produce the same summary statistics in a single line
regimen_df_agg = regimen_df.agg({'Tumor Volume (mm3)': (['mean', 'median', 'var', 'std', 'sem'])})
regimen_df_agg
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
final_timepoint_df = clean_df.groupby(['Mouse ID'])
final_timepoint_df = final_timepoint_df['Timepoint'].max()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
final_merge = pd.merge(clean_df, final_timepoint_df, on=('Mouse ID', 'Timepoint'))
final_merge
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_volume_data = []
drug_1 = []
drug_2 = []
drug_3 = []
drug_4 = []

for drug in treatments:
    tumor_volume_data.append(final_merge.loc[final_merge['Drug Regimen'] == drug, :])
    
drug_1 = tumor_volume_data[0]['Tumor Volume (mm3)']
drug_2 = tumor_volume_data[1]['Tumor Volume (mm3)']
drug_3 = tumor_volume_data[2]['Tumor Volume (mm3)']
drug_4 = tumor_volume_data[3]['Tumor Volume (mm3)']

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in treatments:
    quartiles = final_merge.loc[final_merge['Drug Regimen'] == drug, :]['Tumor Volume (mm3)'].quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    
    print(f"The lower quartile of {drug} is: {lowerq}")
    print(f"The upper quartile of {drug} is: {upperq}")
    print(f"The interquartile range of {drug} is : {iqr}")
    print(f"The median of {drug} is: {quartiles[0.5]}")
    
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f"Values below {lower_bound} could be outliers.")
    print(f"Values above {upper_bound} could be outliers.")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig1, ax1 = plt.subplots()
ax1.set_title('Final Tumor Volume per Drug Regimen')
ax1.set_ylabel('Tumor Volume (mm3)')

ax1.boxplot([drug_1, drug_2, drug_3, drug_4], flierprops={'markerfacecolor': 'g'})
plt.xticks([1,2,3,4], treatments)
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
weight = final_merge.loc[final_merge['Drug Regimen'] == 'Capomulin']
weight_avg = weight.groupby('Mouse ID').mean()
x_values = weight_avg["Weight (g)"]
y_values = weight_avg["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = f"y = {round(slope, 2)} x + {round(intercept, 2)}"
plt.scatter(x_values, y_values)
plt.plot(x_values, regress_values, color='red')
plt.xlabel("Average Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title('Capomulin - Average Mouse Weight Versus Tumor Volume')
plt.show()
```
